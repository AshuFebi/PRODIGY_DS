Importing the necessary Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib as ma 
import seaborn as sb 
import matplotlib.pyplot as plt
```

Reading the CSV File

```python
data = pd.read_csv("./API_SP.POP.TOTL_DS2_en_csv_v2_6508519/population.csv")
```

Create a Bar Chart or Histogram to Visualise the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population 


1.Creating a bar chart to visualize the top 15 countries with the highest populations in the year 2022.


```python
#Extracting the required columns
countries = data['Country Name']
population_2022 = data['2022']

#sorting
sorted_data = data.sort_values(by='2022', ascending=False)

#top 15 countries of high population
top_n = 15
top_countries = sorted_data['Country Name'].head(top_n)

#creating a bar chart
plt.figure(figsize=(12, 7))
plt.bar(top_countries, sorted_data['2022'].head(top_n), color='purple')
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title(f'Distribution of Population in 2022 - Top {top_n} Countries')
plt.xticks(rotation=45)  
plt.show()
```

2.Creating a bar chart to visualize selected countries populations in the year 1996.

```python
# Extracting required columns 
countries = data['Country Name']
population_1996 = data['1996']

# List of countries 
selected_countries = ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea']

# Filtering the DataFrame for the selected countries
selected_data = data[data['Country Name'].isin(selected_countries)]

# Sorting
sorted_data = selected_data.sort_values(by='1996', ascending=False)

# Creating a bar chart
plt.figure(figsize=(11, 6))
plt.bar(sorted_data['Country Name'], sorted_data['1996'], color='pink')
plt.xlabel('Country')
plt.ylabel('Population in 1996')
plt.title(f'Distribution of Population of selected countries in 1996')
plt.xticks(rotation=45)
plt.show()
```

3.Creating a bar chart to visualize populations of India and some of it's neighboring countries in the year 2002.

```python
# Extracting required columns 
countries = data['Country Name']
population_2002 = data['2002']

# List of countries 
selected_countries = ['India', 'Pakistan', 'China', 'Nepal', 'Bhutan', 'Myanmar', 'Vietnam']

# Filtering the DataFrame for the selected countries
selected_data = data[data['Country Name'].isin(selected_countries)]

# Creating a bar chart
plt.figure(figsize=(14, 7))
plt.bar(selected_data['Country Name'], selected_data['2002'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2002')
plt.title(f'Distribution of Population of India and its neighboring countries in 2002')
plt.xticks(rotation=45)
plt.show()
```

4.Histogram for the Population in 2020 across All Countries:

```python
population_2022_all_countries = data['2020']

plt.figure(figsize=(10, 6))
plt.hist(population_2022_all_countries, bins=20, color='orange', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution Across All Countries (2022)')
plt.show()
```
